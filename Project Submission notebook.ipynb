```python
fifa_df = df[df['tournament'] == 'FIFA World Cup']
fifa_df
```

```python
fifa_df.loc[:, 'total_score'] = fifa_df['home_score'] + fifa_df['away_score']
```

```python
fifa_goals_by_country = fifa_df.groupby('country')['total_score'].sum()
fifa_goals_by_country
```

```python
country_highest_fifa_goals = fifa_goals_by_country.idxmax()
country_highest_fifa_goals
```
